```python
import cv2
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from tqdm import tqdm_notebook as tqdm
%matplotlib inline
```

```python
def signature(img_param, n_param):
    img_amelie = img_param
    n = n_param
    seed = []
    for i in range(256//n):
        seed.append(i*n)
    last = seed[-1]
    if last + n <= 255:
        seed.append(last + n)
    dict_cubes = {}
    for i in seed:
        for j in seed:
            for k in seed:
                dict_cubes[(i,j,k)] = 0
    for i in img_amelie:
        for j in i:
            b = j[0]
            g = j[1]
            r = j[2]
            b = b - (b % n)
            g = g - (g % n)
            r = r - (r % n)
            index = (b, g, r)
            dict_cubes[index] = dict_cubes[index] + 1
    img_signature = []
    for i in dict_cubes:
        num = dict_cubes[i]
        for k in range(num):
            img_signature.append([np.uint8(i[0]),np.uint8(i[1]),np.uint8(i[2])])
    img_signature = np.array([img_signature])
    img_signature = cv2.resize(img_signature, (500, 100)) 
    plt.figure(figsize=(15,5))
    plt.imshow(cv2.cvtColor(img_signature, cv2.COLOR_BGR2RGB))
    plt.show()
```

```python
img_klimt_1 = cv2.imread('./imgs/klimt_1.png')
plt.imshow(cv2.cvtColor(img_klimt_1, cv2.COLOR_BGR2RGB))
plt.show()
```
